```python
import matplotlib.pyplot as plt
```

```python
def plot_two_line_graphs(x, y1, y2, label1='F1-Score', label2='Accuracy',x_label="",y_label=""):
    plt.plot(x, y2, marker='s', linestyle='--', color='red', label=label2)
    plt.plot(x, y1, marker='o', linestyle='-', color='orange', label=label1)
    

    for i in range(len(x)):
        plt.text(x[i], y1[i], f' {y1[i]}', fontsize=9, ha='left', color='black')
        plt.text(x[i], y2[i], f'{y2[i]}', fontsize=9, ha='left', color='black')

    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(x_label +' Variation')
    plt.xticks(ticks=x)
    plt.legend()
    plt.grid(True)
    plt.show()
```
